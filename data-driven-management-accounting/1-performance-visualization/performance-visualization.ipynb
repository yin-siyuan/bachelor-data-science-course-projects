# üìä Balanced Scorecard Performance Analysis

This notebook presents an interactive Balanced Scorecard dashboard for GEAR, a premium bicycle manufacturer with markets in North America, Europe, and South America. The visualization enables dynamic filtering and offers performance insights for strategic decision-making.


```python
import pandas as pd
import numpy as np 
import ipywidgets
import pandas as pd
from matplotlib import pyplot as plt
```

## üéØ Dashboard Design & Color Coding Logic

We developed a user-friendly interface to display performance KPIs per market. The dashboard uses the following color scheme:

- ‚úÖ Green: Target met or exceeded
- üüß Orange: Within 10% below target
- üî¥ Red: More than 10% below target

This intuitive design allows managers to quickly spot underperformance and take corrective action.

```python
df = pd.read_excel(r'../data/performance-visualization.xlsx')
```

```python
# (Optional) Disable chained assignment warning ‚Äî but better to use .loc to avoid it altogether
pd.options.mode.chained_assignment = None  

# Calculate the difference between target and actual performance
df['Target Diff'] = df['Target Performance'] - df['Actual Performance']

# Initialize a new column to store target achievement status
df['Target Reached'] = ""

# Define positive-oriented measures: higher values indicate better performance
positive_measures = [
    'Profit Margin', 'Sales Growth', 'Customization Upcharge Revenue',
    'Customer Satisfaction Score', 'Net Promoter Score', 'Customer Retention Rate',
    'Customization Orders', 'Employee Training Hours', 'Employee Satisfaction Score', 
    'Innovation Index'
]

# Define negative-oriented measures: lower values indicate better performance
negative_measures = [
    "Order Fulfillment Time", 'Product Defect Rate',
    'Customization Error Rate', 'Efficiency of Customization Process'
]

# Iterate through each row and determine if the performance target was achieved
for i in range(len(df)):
    measure = df.loc[i, 'Measure']                  # Name of the current performance measure
    diff = df.loc[i, 'Target Diff']                 # Difference between target and actual value
    target = df.loc[i, 'Target Performance']        # Target value for the current measure

    # Evaluation logic for positive-oriented measures
    if measure in positive_measures:
        if diff <= 0:
            df.loc[i, 'Target Reached'] = "yes"     # Actual ‚â• Target: goal achieved
        elif diff <= 0.1 * target:
            df.loc[i, 'Target Reached'] = "missed by no more than 10%"  # Slightly missed
        else:
            df.loc[i, 'Target Reached'] = "no"      # Missed target by more than 10%

    # Evaluation logic for negative-oriented measures
    elif measure in negative_measures:
        if diff >= 0:
            df.loc[i, 'Target Reached'] = "yes"     # Actual ‚â§ Target: goal achieved
        elif diff >= -0.1 * target:
            df.loc[i, 'Target Reached'] = "missed by no more than 10%"  # Slightly missed
        else:
            df.loc[i, 'Target Reached'] = "no"      # Missed target by more than 10%
```

```python
def reached_target(row):
    # Define color styles for different target achievement statuses
    fail = 'background-color: red;'          # For targets that were missed by more than 10%
    exceed = 'background-color: lawngreen;'  # For targets that were fully achieved
    almost = 'background-color: orange;'     # For targets that were narrowly missed (within 10%)
    default = ''                              # No styling for non-target columns

    # Return a list of style strings, one for each column being styled in the DataFrame

    if row['Target Reached'] == "yes":
        return [default, exceed]              # Highlight the 'Target Reached' cell in green
    elif row['Target Reached'] == "missed by no more than 10%":
        return [default, almost]              # Highlight in orange for near-misses
    else:
        return [default, fail]                # Highlight in red for full misses
```

```python
# Create a dropdown widget to select a market from the unique values in the 'Market' column
drop_down = ipywidgets.Dropdown(
    options=list(df['Market'].unique()),  # Populate dropdown with unique market names
    value='North America',                # Set default selected value
    description='Market:',                # Label shown next to dropdown
    disabled=False                        # Widget is active (not disabled)
)

# Define a function to generate a styled balanced scorecard for the selected market
def balanced_scorecard(market):
    # Filter the DataFrame to only include rows for the selected market
    kpifilter = df[df['Market'] == market]

    # Select and reorder the relevant columns to display in the scorecard
    kpifilter = kpifilter[['Perspective', 'Measure', 'Definition',
                           'Target Performance', 'Actual Performance', 'Target Reached']]

    # Apply conditional formatting using the reached_target() function
    # Style is applied only to the 'Actual Performance' and 'Target Reached' columns
    return kpifilter.style.apply(reached_target, subset=['Actual Performance', 'Target Reached'], axis='columns')

# Create an interactive widget that calls the balanced_scorecard function
# It updates automatically when the dropdown value changes
ipywidgets.interact(balanced_scorecard, market=drop_down)
```

## üîç Strategic Measure Analysis

GEAR follows a **Product Differentiation** strategy. The following Balanced Scorecard measures were assessed for their strategic impact:

- **Customization Upcharge Revenue (Financial Perspective)**  
  This measure exceeded targets in North America and Europe, suggesting high willingness to pay for premium options. Underperformance in South America may indicate pricing sensitivity.

- **Net Promoter Score (Customer Perspective)**

- **Product Defect Rate (Internal Business Process)**  
  Targets were missed in all markets, potentially eroding brand perception. This insight informs the next phase of the project: predictive modeling.

## üåç Recommendation: Integrating Asian and Australian Markets into GEAR's Balanced Scorecard

To support GEAR‚Äôs expansion into the Asian and Australian markets, the Balanced Scorecard (BSC) framework should be adapted with both consistency and flexibility in mind:

- **Internal Business Process** and **Learning and Growth** perspectives  
  These perspectives focus on operational efficiency, innovation, and employee development, which are foundational across all markets. Therefore, the current performance measures and targets can remain largely unchanged when applied to new regions.

- **Financial** and **Customer** perspectives  
  These metrics are more sensitive to market-specific factors such as customer familiarity, brand recognition, and local competition. For these perspectives, it is recommended to **temporarily lower performance targets** to account for the initial market entry phase. This includes adjusting expectations for:
  
  - Profit margin  
  - Customization upcharge revenue  
  - Customization order volume  

This approach ensures that short-term market realities are reflected in performance evaluation, while long-term strategic alignment is preserved across regions. It provides a more realistic basis for decision-making and performance monitoring during the ramp-up phase.

## üìà Enhanced Visualization (Inspired by a [case study](https://www.sciencedirect.com/science/article/abs/pii/S036136822100074X))


We extended the dashboard with market-level scorecard tiles, combining categorical evaluation with graphical summaries. This visual format reflects advanced practices observed in certain industry executive dashboards, supporting improved readability and strategic decision-making.

```python
df4 = pd.read_excel(r'../data/performance-visualization.xlsx')

LOWER_BETTER_MEASURES = [
    'Order Fulfillment Time',
    'Product Defect Rate',
    'Customization Error Rate',
    'Efficiency of Customization Process'
]

def get_color(actual, target, measure):
    if measure in LOWER_BETTER_MEASURES:
        if actual <= target:
            return 'green'
        elif actual <= 1.1 * target:
            return 'orange'
        else:
            return 'red'
    else:
        if actual >= target:
            return 'green'
        elif actual >= 0.9 * target: 
            return 'orange'
        else:
            return 'red'

def calculate_ratio(actual, target, measure):
    if measure in LOWER_BETTER_MEASURES:
        return target / actual if actual != 0 else 0  
    else:
        return actual / target if target != 0 else 0  

markets = df4['Market'].unique()

for market in markets:
    market_data = df4[df4['Market'] == market].copy()
    
    numeric_cols = ['Target Performance', 'Actual Performance']
    market_data[numeric_cols] = market_data[numeric_cols].apply(pd.to_numeric, errors='coerce')
    
    market_data['Performance Ratio'] = market_data.apply(
        lambda x: calculate_ratio(x['Actual Performance'], x['Target Performance'], x['Measure']), 
        axis=1
    )
    
    market_data['Color'] = market_data.apply(
        lambda x: get_color(x['Actual Performance'], x['Target Performance'], x['Measure']), 
        axis=1
    )
    
    def abbreviate(phrase):
        return ''.join(word[0].upper() for word in phrase.split())

    market_data['Abbreviation'] = market_data['Measure'].apply(abbreviate)
    
    fig, ax = plt.subplots(figsize=(12, 8))
    
    abbreviations = market_data['Abbreviation'].values
    
    y_pos = np.arange(len(market_data))
    
    ax.plot([0.9] * len(y_pos), y_pos, 'r--', linewidth=3, alpha=0.5)
    ax.plot([1]*len(y_pos), y_pos, 'g--', linewidth=3, alpha=0.5) 
    ax.plot(market_data['Performance Ratio'], y_pos, 'o-', color='black', linewidth=2, markersize=8)
    
    for i, (color, ratio) in enumerate(zip(market_data['Color'], market_data['Performance Ratio'])):
        ax.plot(ratio, i, 'o', color=color, markersize=12)
    
    ax.set_yticks(y_pos)
    ax.set_yticklabels(abbreviations, fontsize=10)
    ax.set_xlim(0.65, 1.45) 
    ax.set_xticks([]) 
    ax.set_title(f'{market}', fontsize=20, pad=20)
    ax.grid(True, axis='x', linestyle='--', alpha=0.6)
    
    plt.tight_layout()
    plt.show()
```

## ‚öñÔ∏è Evaluation of Supplementary Graphical Visualization for Market Performance

The proposed multimodal visualization leverages multiple worm diagrams (also referred to as bullet graphs) to support comparative performance analysis across different markets.

**Advantages**

- **Improved Accessibility**  
  The graphical format enables stakeholders‚Äîparticularly non-technical users such as executives and line managers‚Äîto quickly grasp key performance outcomes without requiring in-depth analytical expertise.

- **Facilitated Comparison**  
  The layout supports intuitive, side-by-side evaluation of performance across multiple markets, streamlining the identification of underperforming or outperforming regions.

- **Pattern Recognition**  
  Visual emphasis on trends and deviations makes it easier to detect emerging patterns or anomalies, which might be overlooked in tabular or purely numerical formats.

**Challenges**

- **Risk of Oversimplification**  
  The abstraction of data into visual elements can lead to important numerical subtleties being overlooked, potentially reducing analytical precision.

- **Potential for Misinterpretation**  
  Without appropriate context or supplementary metrics, the visualization may unintentionally convey a misleading impression of performance‚Äîespecially if graphical scaling or thresholds are not clearly defined.

Overall, while the visualization offers strong benefits for strategic communication and high-level performance monitoring, it should be complemented by detailed data views to preserve analytical rigor.
